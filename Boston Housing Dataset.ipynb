```python
#the Full code to the FIT() model is:
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = boston[['RM']]
Y = boston['MEDV']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
  test_size = 0.3,
  random_state=1)
model = LinearRegression()
model.fit(X_train, Y_train)
model.intercept_.round(2)
model.coef_.round(2)
```

```python
import numpy as np
new_RM=np.array([6.5]).reshape(-1,1)
model.predict(new_RM)
```

```python
model.intercept_.round(2)+model.coef_.round(2)*6.5
```

```python
y_test_predicted=model.predict(X_test)
y_test_predicted.shape
type(y_test_predicted)
```

```python
Y_test.shape
```

```python
#examine model performance by visually comparing the fitted line and the true observations in the test set.
plt.scatter(X_test, Y_test,
           label='testing data');
plt.plot(X_test, y_test_predicted,
        label='prediction', linewidth=3)
plt.xlabel('RM'); plt.ylabel('MEDV')
plt.legend(loc='upper left')
plt.show()
```

```python
residuals=Y_test-y_test_predicted
#plot the residuals
plt.scatter(X_test, residuals)
#plot the horizontal line
plt.hlines(y=0,
          xmin=X_test.min(), xmax=X_test.max(),
          linestyle='--')
#set xlim
plt.xlim((4.9))
plt.xlabel('RM')
plt.ylabel('residual')
plt.show()

```
